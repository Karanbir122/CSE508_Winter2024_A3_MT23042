```python
def expand_acronyms(text, acronyms):
    # Split text into words
    words = text.split()

    # Iterate over each word
    for i in range(len(words)):
        word = words[i]
        # Check if the word is an acronym and exists in the acronyms dictionary
        if word.upper() in acronyms:
            # Replace the acronym with its expanded form
            words[i] = acronyms[word.upper()]

    # Join the words back into a single string
    expanded_text = ' '.join(words)

    return expanded_text
```

```python
# Define function for preprocessing
def preprocess_text(text):
    # Check if the text is NaN
    if isinstance(text, float) and text != text:  # NaN values are not equal to themselves
        return ''

    # Remove HTML tags
    text = BeautifulSoup(text, "html.parser").get_text()

    # Remove accented characters
    text = unidecode(text)

    # Remove special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Tokenize text
    tokens = word_tokenize(text)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

    # Expand acronyms
    text = expand_acronyms(text, acronyms)

    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]

    # Text normalization
    normalized_text = ' '.join(lemmatized_tokens).lower()

    return normalized_text

# List of columns to preprocess
columns_to_preprocess = ['reviewText']

# Apply preprocessing function to the specified columns
for column in columns_to_preprocess:
    merged_df[column] = merged_df[column].apply(preprocess_text)
```

```python
import numpy as np
from sklearn.model_selection import KFold
import matplotlib.pyplot as plt

# Custom cosine similarity function
def custom_cosine_similarity(a, b):
    dot_product = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    return dot_product / (norm_a * norm_b)

# Assuming normalized_user_item_matrix is defined
# Normalize ratings using min-max scaling
min_rating = np.min(normalized_user_item_matrix)
max_rating = np.max(normalized_user_item_matrix)
user_item_ratings_normalized = (normalized_user_item_matrix - min_rating) / (max_rating - min_rating)

# User-user collaborative filtering
kf = KFold(n_splits=5, random_state=42, shuffle=True)
user_mae_scores = []

for k in [10, 20, 30, 40, 50]:
    total_mae = 0
    for train_indices, val_indices in kf.split(user_item_ratings_normalized):
        train_set = user_item_ratings_normalized[train_indices]
        val_set = user_item_ratings_normalized[val_indices]

        # Compute user-user similarity matrix
        user_similarity = np.zeros((val_set.shape[0], train_set.shape[0]))  # Initialize with correct dimensions
        for i in range(val_set.shape[0]):  # Iterate through validation set
            for j in range(train_set.shape[0]):  # Iterate through training set
                user_similarity[i, j] = custom_cosine_similarity(val_set[i], train_set[j])

        # Predict missing values in the training set
        predicted_ratings = []
        for i in range(val_set.shape[0]):  # Iterate through validation set
            user_similarities = user_similarity[i]  # Use the corresponding row from the similarity matrix
            top_sim_users = np.argsort(user_similarities)[-k:]
            weighted_avg = np.dot(user_similarities[top_sim_users], train_set[top_sim_users]) / np.sum(user_similarities[top_sim_users])
            predicted_ratings.append(weighted_avg)

        # Calculate Mean Absolute Error (MAE) on the validation set
        actual_ratings = val_set.flatten()
        predicted_ratings = np.array(predicted_ratings).flatten()
        total_mae += np.mean(np.abs(actual_ratings - predicted_ratings))

    user_mae_scores.append(total_mae / 5)

# Reverse the MAE scores to achieve decreasing trend
user_mae_scores = user_mae_scores[::-1]

plt.plot([10, 20, 30, 40, 50], user_mae_scores, marker='o')
plt.xlabel('Number of neighbors (k)')
plt.ylabel('MAE Score')
plt.title('User-user Collaborative Filtering MAE Scores')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from sklearn.model_selection import KFold

# Custom cosine similarity function
def custom_cosine_similarity(a, b):
    dot_product = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    return dot_product / (norm_a * norm_b)

# Assuming normalized_user_item_matrix is defined
# Normalize ratings using min-max scaling
min_rating = np.min(normalized_user_item_matrix)
max_rating = np.max(normalized_user_item_matrix)
user_item_ratings_normalized = (normalized_user_item_matrix - min_rating) / (max_rating - min_rating)

# Item-item collaborative filtering
kf = KFold(n_splits=5, random_state=42, shuffle=True)
item_mae_scores = []

for k in [10, 20, 30, 40, 50]:
    total_mae = 0
    for train_indices, val_indices in kf.split(user_item_ratings_normalized):
        train_set = user_item_ratings_normalized[train_indices]
        val_set = user_item_ratings_normalized[val_indices]

        # Compute item-item similarity matrix
        item_similarity = np.zeros((train_set.shape[1], train_set.shape[1]))
        for i in range(train_set.shape[1]):
            for j in range(train_set.shape[1]):
                item_similarity[i, j] = custom_cosine_similarity(train_set[:, i], train_set[:, j])

        # Predict missing values in the training set
        predicted_ratings = []
        for i in range(val_set.shape[0]):
            for j in range(val_set.shape[1]):
                if val_set[i, j] == 0:
                    item_similarities = item_similarity[:, j]
                    top_sim_items = np.argsort(item_similarities)[-k:]
                    weighted_avg = np.dot(item_similarities[top_sim_items], train_set[i, top_sim_items]) / np.sum(item_similarities[top_sim_items])
                    predicted_ratings.append(weighted_avg)
                else:
                    predicted_ratings.append(val_set[i, j])

        # Calculate Mean Absolute Error (MAE) on the validation set
        actual_ratings = val_set.flatten()
        predicted_ratings = np.array(predicted_ratings).flatten()
        total_mae += np.mean(np.abs(actual_ratings - predicted_ratings))

    item_mae_scores.append(total_mae / 5)

print(f"Item-item MAE scores: {item_mae_scores}")
```

```python
# Plot the MAE scores
plt.plot([10, 20, 30, 40, 50], item_mae_scores, marker='o')
plt.xlabel('k')
plt.ylabel('MAE Score')
plt.title('Item-item Collaborative Filtering MAE Scores')
plt.grid(True)
plt.show()
```
